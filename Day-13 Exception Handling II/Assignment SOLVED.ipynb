Custom exceptions should also inherit from the Exception class to ensure that they behave like other exceptions in the language, can be used in exception handling code, and are consistent with Python's conventions for exceptions. Inheriting from the Exception class provides our custom exception with all of the functionality and behavior of the built-in exceptions, making our code easier to understand and maintain.

**Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.**

The LookupError class is a built-in exception class in Python that is raised when a lookup or indexing operation fails. It is the base class for several more specific lookup and indexing exception classes, including KeyError, IndexError, and AttributeError. Here are two examples of specific exceptions that are defined in the LookupError class:
- **KeyError**: This exception is raised when a dictionary key is not found in the dictionary. For example, consider the following code:

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}
value = my_dict['d']  # This will raise a KeyError
```

- **IndexError**: This exception is raised when attempting to access an index in a sequence that is out of range. For example, consider the following code:

```python
my_list = [1, 2, 3]
value = my_list[3]  # This will raise an IndexError
```

**Q5. Explain ImportError. What is ModuleNotFoundError?**

**ImportError** is a built-in exception class in Python that is raised when an imported module or module attribute is not found. This can occur for several reasons, such as a misspelled module name or an improperly installed module. The ImportError class is a base class for several more specific import exceptions, such as ModuleNotFoundError.

**ModuleNotFoundError** is a more specific exception class that was introduced in Python 3.6. It is raised when an imported module is not found. This exception is raised when a module that is specified in an import statement cannot be found in any of the locations on the sys.path list.
