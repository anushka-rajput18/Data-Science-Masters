**Q1. List any five functions of the pandas library with execution.**

here are five common functions of the pandas library with examples:
1. **read_csv()**: Reads a CSV file and returns a DataFrame object.

```python
df = pd.read_csv("players.csv")
```

2. **head()**: Returns the first n rows of a DataFrame.

```python
df.head()
```

3. **groupby()**: Groups the DataFrame by a column or columns and applies a function to each group.

```python
grouped = df.groupby('Age')
grouped.first()
```

3. **len()**: length of the dataframe

```python
len(df)
```

4. **unique()**: Returns the unique values in a column of a DataFrame.

```python
unique_values = df['Tm'].unique()
print(unique_values)
```

5. **sort_values()**: Sorts the rows of a DataFrame by one or more columns.

```python
sorted_df = df.sort_values(["Age"], ascending=[True])
print(sorted_df)
```

**Q.2 Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.**

```python
def reindex_dataframe(df):
    # Get the number of rows in the DataFrame
    num_rows = df.shape[0]
    
    # Create a new index that starts from 1 and increments by 2
    new_index = pd.Index(range(1, num_rows*2, 2))
    
    # Re-index the DataFrame
    df_reindexed = df.set_index(new_index)
    
    return df_reindexed
```

```python
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
df_reindexed = reindex_dataframe(df1)
print(df_reindexed)
```
